```python
df = sns.load_dataset('tips')
```

```python
df
```

```python
sns.relplot(x = 'total_bill',y = 'tip',data = df)
```

```python
sns.scatterplot(x = df.total_bill,y = df.tip,data = df,hue = 'smoker')
```

```python
sns.relplot(x = 'total_bill',y = 'tip',data = df,hue = 'smoker')
```

```python
sns.relplot(x = 'total_bill',y = 'tip',data = df,hue = 'size')
```

```python
sns.relplot(x = 'total_bill',y='tip',style = 'smoker',data=df)
```

```python
sns.relplot(x = 'total_bill',y = 'tip',hue = 'smoker',style = 'time',data = df)
```

```python
#relplot to make a line plot>> we generated a data
time = np.arange(500)
value = np.random.randn(500)
df = pd.DataFrame({'time':time,'value':value})
```

```python
df
```

```python
sns.relplot(x = 'time',y = 'value',kind = 'line',sort = True,data = df)
```
